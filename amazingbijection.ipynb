```python
def findavoid(n,pt, d):
    pattern = []
    code = []
    for p in Permutations(n):
        if p.avoids(pt):
            pattern.append(p)
            code.append(p.to_lehmer_code())
    if d=='pa':
        return pattern
    if d=='c':
        return code
    if d=='l':
        return len(pattern)
```

```python
def code_inversion_count(code):
    n = len(code)
    count = 0
    for j in range(n):
        for k in range(j + 1, n):
            if code[j] > code[k]:
                count += 1
    return count

def bw_partition_counts_by_parts(n):
    counts = [0] * (n + 1)
    all_parts = []
    all_list = []

    def backtrack(b_left, w_left, last, cur):
        if b_left == 0 and w_left == 0:
            counts[len(cur)] += 1
            all_parts.append(cur[:])
            return

        for b in range(b_left, 0, -1):
            for w in range(w_left, -1, -1):
                part = (b, w)
                if last is None or part <= last:
                    cur.append(part)
                    backtrack(b_left - b, w_left - w, part, cur)
                    cur.pop()

    for total_b in range(1, n + 1):
        total_w = n - total_b
        backtrack(total_b, total_w, None, [])

    for p in all_parts:
        pbw = ["b" * b + "w" * w for (b, w) in p]
        all_list.append(pbw)

    # print("\ncounts by number of parts:")
    # for k in range(1, n + 1):
    #     print(f"{k} part(s): {counts[k]}")

    return counts, all_list
    
def bw_strings(lst):
    result = []
    
    for i in range(len(lst)):
        x = lst[i]
        if x != 0:
            inv_count = sum(1 for y in lst[i+1:] if y < x)
            s = "b" * x + "w" * (inv_count - x)
            result.append(s)
    
    result.sort(key=lambda s: (s.count('b'), len(s)), reverse=True)
    return result

def codetobw_all(n):
    lst = []
    for p in Permutations(n+1):
        c= p.to_lehmer_code()
        if code_inversion_count(c) == n:
            lst.append(c)
    maplist = []
    for code in lst:
        maplist.append(bw_strings(code))
    return(maplist)

def fixinvall(n):
    lst = []
    for p in Permutations(n+1):
        c= p.to_lehmer_code()
        if code_inversion_count(c) == n:
            lst.append(c)
    return lst
```

```python
def part_code(part):
    k = part.count('b')
    n = len(part)
    return [k] + [0] * n


def combine_codes(code1, code2):
    A = code1[:-1]
    B = code2[:-1]
    jump = sum(1 for a in A for b in B if a > b)
    merged = A + B
    first_nonzero = merged[0]
    D = merged[1:]
    if jump > len(D):
        raise ValueError("jump number larger then D")
    for i in D[:jump]:
        if i == first_nonzero:
            raise ValueError("stupid jump")
    new_code = D[:jump] + [first_nonzero] + D[jump:] + [0]
    return new_code

def bw_parts_to_code(parts):
    if isinstance(parts, str):
        return part_code(parts)

    if len(parts) == 0:
        return []

    if len(parts) == 1:
        return part_code(parts[0])

    current = part_code(parts[-1])
    for part in reversed(parts[:-1]):
        current = combine_codes(part_code(part), current)

    return current
```

```python
def code_lehmer(code):
    n = len(code)
    return [sum(1 for j in range(i+1, n) if code[j] < code[i]) for i in range(n)]
```
