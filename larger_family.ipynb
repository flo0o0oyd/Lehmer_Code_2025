```python
import sage.combinat.permutation as permutation
```

```python
def get_lehmer(n,pt):
    dw = DyckWords(n)
    lst = []
    for d in dw:
        if pt == a123:
            lst.append(dw_alpha_last(d))
        if pt == a132:
            lst.append(dw_beta_last(d))
        if pt == a213:
            lst.append(dw_gamma_last(d))
        if pt == a231:
            lst.append(dw_delta_last(d))
        if pt == a312:
            lst.append(dw_epsilon_last(d))
        if pt == a321:
            lst.append(dw_zeta_last(d))
    return lst

def get_perm(n,pt):
    dw = DyckWords(n)
    lst = []
    for d in dw:
        if pt == a123:
            lst.append(permutation.from_lehmer_code(dw_alpha_last(d)))
        if pt == a132:
            lst.append(permutation.from_lehmer_code(dw_beta_last(d)))
        if pt == a213:
            lst.append(permutation.from_lehmer_code(dw_gamma_last(d)))
        if pt == a231:
            lst.append(permutation.from_lehmer_code(dw_delta_last(d)))
        if pt == a312:
            lst.append(permutation.from_lehmer_code(dw_epsilon_last(d)))
        if pt == a321:
            lst.append(permutation.from_lehmer_code(dw_zeta_last(d)))
    return lst
        

def first_nonzero_index(lst):
    for i in range(len(lst)):
        if lst[i] != 0:
            return i
    return None

a123 = [1,2,3]
a132 = [1,3,2]
a213 = [2,1,3]
a231 = [2,3,1]
a312 = [3,1,2]
a321 = [3,2,1]
```

```python
def standardize3(a, b, c):
    vals = [a, b, c]
    sorted_vals = sorted(vals)
    rank = {}
    for i, v in enumerate(sorted_vals):
        rank[v] = i + 1
    return [rank[a], rank[b], rank[c]]

def code_inversion_count(code):
    n = len(code)
    count = 0
    for j in range(n):
        for k in range(j + 1, n):
            if code[j] > code[k]:
                count += 1
    return count
```

```python
def contains_pattern_3(perm, pattern, kind="1-2-3"):
    pat = [int(ch) for ch in pattern]
    n = len(perm)

    if kind == "123":
        for i in range(n - 2):
            if standardize3(perm[i], perm[i+1], perm[i+2]) == pat:
                return True
        return False

    elif kind == "12-3":
        for i in range(n - 2):
            for k in range(i + 2, n):
                if standardize3(perm[i], perm[i+1], perm[k]) == pat:
                    return True
        return False

    elif kind == "1-23":
        for i in range(n - 2):
            for j in range(i + 1, n - 1):
                if standardize3(perm[i], perm[j], perm[j+1]) == pat:
                    return True
        return False

    elif kind == "1-2-3":
        for i in range(n):
            for j in range(i + 1, n):
                for k in range(j + 1, n):
                    if standardize3(perm[i], perm[j], perm[k]) == pat:
                        return True
        return False

    else:
        raise ValueError("kind must be one of '123', '1-23', '12-3', '1-2-3'")

def avoids_pattern_3(perm, pattern, kind="1-2-3"):
    return not contains_pattern_3(perm, pattern, kind)

def all_avoiding_perms(n, pattern, kind="1-2-3"):
    good = []
    for p in Permutations(n):
        p = list(p)
        if avoids_pattern_3(p, pattern, kind):
            good.append(p)
    return good

def all_avoiding_codes(n, pattern, kind="1-2-3"):
    code_list = []
    for p in Permutations(n):
        p_list = list(p)
        if avoids_pattern_3(p_list, pattern, kind):
            code_list.append(list(Permutation(p_list).to_lehmer_code()))
    return code_list
```
