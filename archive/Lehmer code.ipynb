```python
a123 = [1,2,3]
a132 = [1,3,2]
a213 = [2,1,3]
a231 = [2,3,1]
a312 = [3,1,2]
a321 = [3,2,1]

def findavoid(n,pt, d):
    pattern = []
    code = []
    for p in Permutations(n):
        if p.avoids(pt):
            pattern.append(p)
            code.append(p.to_lehmer_code())
    if d=='pa':
        return pattern
    if d=='c':
        return code
    if d=='l':
        return len(pattern)

def findpat(n,pt,d):
    pattern = []
    code = []
    for p in Permutations(n):
        if p.has_pattern([1,2,3]):
            pattern.append(p)
            code.append(p.to_lehmer_code())
    if d=="pa":
        return pattern
    if d=='c':
        return code
    if d=='l':
        return len(pattern)


def checkhas123weak(seq):
    n = len(seq)
    for j in range(n):
        for k in range(j):
            for i in range(k):
                if seq[i] <= seq[k] <= seq[j]:
                    return True
    return False

def checkhas132weak(seq):
    n = len(seq)
    for j in range(n):
        for k in range(j):
            for i in range(k):
                if seq[i] <= seq[j] <= seq[k]:
                    return True
    return False

def checkhas213weak(seq):
    n = len(seq)
    for j in range(n):
        for k in range(j):
            for i in range(k):
                if seq[k] <= seq[i] <= seq[j]:
                    return True
    return False

def checkhas231weak(seq):
    n = len(seq)
    for j in range(n):
        for k in range(j):
            for i in range(k):
                if seq[j] <= seq[i] <= seq[k]:
                    return True
    return False

def checkhas312weak(seq):
    n = len(seq)
    for j in range(n):
        for k in range(j):
            for i in range(k):
                if seq[k] <= seq[j] <= seq[i]:
                    return True
    return False

def checkhas321weak(seq):
    n = len(seq)
    for j in range(n):
        for k in range(j):
            for i in range(k):
                if seq[j] <= seq[k] <= seq[i]:
                    return True
    return False


def checkhas(seq,checkpt):
    n = len(seq)
    if checkpt == "123":
        for j in range(n):
            for k in range(j):
                for i in range(k):
                    if seq[i] < seq[k] < seq[j]:
                        return True
    if checkpt == "132":
        for j in range(n):
            for k in range(j):
                for i in range(k):
                    if seq[i] < seq[j] < seq[k]:
                        return True
    if checkpt =="213":
        for j in range(n):
            for k in range(j):
                for i in range(k):
                    if seq[k] < seq[i] < seq[j]:
                        return True
    if checkpt == "231":
        for j in range(n):
            for k in range(j):
                for i in range(k):
                    if seq[j] < seq[i] < seq[k]:
                        return True
    if checkpt == "312":
        for j in range(n):
            for k in range(j):
                for i in range(k):
                    if seq[k] < seq[j] < seq[i]:
                        return True
    if checkpt == "321":
        for j in range(n):
            for k in range(j):
                for i in range(k):
                    if seq[j] < seq[k] < seq[i]:
                        return True
    if checkpt == "011":
        for j in range(n):
            for k in range(j):
                for i in range(k):
                    if seq[i] < seq[k] == seq[j]:
                        return True
    if checkpt == "001":
        for j in range(n):
            for k in range(j):
                for i in range(k):
                    if seq[i] == seq[k] < seq[j]:
                        return True
    if checkpt == "010":
        for j in range(n):
            for k in range(j):
                for i in range(k):
                    if seq[i] == seq[j] < seq[k]:
                        return True
    if checkpt == "100":
        for j in range(n):
            for k in range(j):
                for i in range(k):
                    if seq[k] == seq[j] < seq[i]:
                        return True
    if checkpt == "110":
        for j in range(n):
            for k in range(j):
                for i in range(k):
                    if seq[j] < seq[k] == seq[i]:
                        return True
    if checkpt == "101":
        for j in range(n):
            for k in range(j):
                for i in range(k):
                    if seq[k] < seq[j] == seq[i]:
                        return True
    if checkpt == "01":
        for j in range(n):
            for k in range(j):
                if seq[k] < seq[j]:
                    return True
    if checkpt == "10":
        for j in range(n):
            for k in range(j):
                if seq[j] < seq[k]:
                    return True
    return False
```

```python
def lehmerM(a):
    n = len(a)
    M = Matrix(ZZ, n, n) 
    for i in range(n):
        for j in range(i, n): 
            count = 0
            for k in range(i+1, j+1):
                if a[i] > a[k]:
                    count += 1
            M[i, j] = count
    return M

def lehmerMrev(a):
    n = len(a)
    M = Matrix(ZZ, n, n)
    for i in range(n):
        for j in range(i, n):
            count = 0
            for k in range(i, j):
                if a[k] > a[j]:
                    count += 1
            M[i, j] = count
    return M

a = [3,1,5,2,4]
print(lehmerM(a))
b = Permutation([1, 4, 2, 3])
b2 = Permutation([3, 1, 2])
b3 = Permutation([1, 2])
print(b.to_lehmer_code())
print(b2.to_lehmer_code())
print(b3.to_lehmer_code())
```
