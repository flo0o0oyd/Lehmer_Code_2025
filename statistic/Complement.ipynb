```python
a123 = [1,2,3]
a132 = [1,3,2]
a213 = [2,1,3]
a231 = [2,3,1]
a312 = [3,1,2]
a321 = [3,2,1]

def findavoid(n,pt, d='c'):
    pattern = []
    code = []
    for p in Permutations(n):
        if p.avoids(pt):
            pattern.append(p)
            code.append(p.to_lehmer_code())
    if d=='pa':
        return pattern
    if d=='c':
        return code
    if d=='l':
        return len(pattern)
```

```python
def findcomple(n,pt):
    pc = []
    for p in findavoid(n,pt,"c"):
        pmax = list(range(n-1, -1, -1))
        p1 = [x - y for x, y in zip(pmax, p)]
        pc.append(p1)
    return pc
```

```python
list1 = findcomple(5,a123)
list2 = findavoid(5,a321)
diff1 = [item for item in list1 if item not in list2]
diff2 = [item for item in list2 if item not in list1]
print(diff1)
print(diff2)
```

```python
def func1(p):
    p = Permutation(p)
    n = len(p)
    c = p.to_lehmer_code()
    cmax = list(range(n - 1, -1, -1))
    c1 = [x - y for x, y in zip(cmax, c)]
    available = list(range(1, n + 1))
    result = []
    for v in c1:
        result.append(available.pop(v))
        
    return Permutation(result)
```

```python
func1([3,4,2,1])
```

```python
def func2(p):
    n = len(p)
    lst = [n+1] * n
    return [a - b for a, b in zip(lst, p)]
```

```python
func2([3,4,2,1])
```

```python
pattern = []
for p in Permutations(5):
    if p.avoids(a123):
        pattern.append(p)
all_equal = all(func1(p) == func2(p) for p in pattern)
print(all_equal)
```
