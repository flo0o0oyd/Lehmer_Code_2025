```python
def is_flush(c):
    n = len(c)
    for i in range(n - 1):
        val = c[i]
        if val == n - i - 1:
            if all(c[j] >= val for j in range(i)):
                return True
    return False
```

```python
a123 = [1,2,3]
a132 = [1,3,2]
a213 = [2,1,3]
a231 = [2,3,1]
a312 = [3,1,2]
a321 = [3,2,1]

def findavoid(n,pt, d='c'):
    pattern = []
    code = []
    for p in Permutations(n):
        if p.avoids(pt):
            pattern.append(p)
            code.append(p.to_lehmer_code())
    if d=='pa':
        return pattern
    if d=='c':
        return code
    if d=='l':
        return len(pattern)

def findpat(n,pt,d):
    pattern = []
    code = []
    for p in Permutations(n):
        if p.has_pattern([1,2,3]):
            pattern.append(p)
            code.append(p.to_lehmer_code())
    if d=="pa":
        return pattern
    if d=='c':
        return code
    if d=='l':
        return len(pattern)
```

```python
def count_flush_codes_all(n):
    count = 0
    for p in Permutations(n):
        c = p.to_lehmer_code()
        if is_flush(c):
            count += 1
    return count
```

```python
for n in range(1, 8):
    print(f"n = {n}: flush codes = {count_flush_codes_all(n)}")
#Number of reducible permutations.
```

```python
def count_flush_codes(n,pt):
    count = 0
    for p in findavoid(n,pt,'pa'):
        c = p.to_lehmer_code()
        if is_flush(c):
            count += 1
    return count
```

```python
for n in range(1, 8):
    print(f"n = {n}: flush codes = {count_flush_codes(n,a321)}")
```
