```python
def print_red(text):
    print("\033[91m" + str(text) + "\033[0m")

def print_green(text):
    print("\033[92m" + str(text) + "\033[0m")

def print_yellow(text):
    print("\033[93m" + str(text) + "\033[0m")

def print_blue(text):
    print("\033[94m" + str(text) + "\033[0m")
```

```python
def check123property(code):
    n = len(code)
    for i in range(n):
        for j in range(i + 1, n):
            if code[i] <= code[j]:
                if code[j] != n - 1 - j:
                    return False
    return True


def check123property_strict(code):
    n = len(code)
    for i in range(n):
        for j in range(i + 1, n):
            if code[i] < code[j]:
                if code[j] != n - 1 - j:
                    return False
    return True


def check123property_adjacent(code):
    n = len(code)
    for i in range(n - 1):
        j = i + 1
        if code[i] <= code[j]:
            if code[j] != n - 1 - j:
                return False
    return True
```

```python
def countNpermutation(n, pa, check_fn, label=""):
    code = []
    for p in Permutations(n):
        code.append(p.to_lehmer_code())
    count = 0
    for c in code:
        if check_fn(c):
            count += 1
    print(f"For n = {n}, {pa} avoiding permutations with {label}: {count}")
    return count
```

```python
for i in range(1, 7):
    countNpermutation(i, a123, check123property, "r(i) <= r(j), all i < j")
for i in range(1, 7):
    countNpermutation(i, a123, check123property_strict, "r(i) < r(j), all i < j")
for i in range(1, 8):
    countNpermutation(i, a123, check123property_adjacent, "r(i) <= r(j), j = i + 1")
```
