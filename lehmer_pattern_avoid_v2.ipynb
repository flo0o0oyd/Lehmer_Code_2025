```python
# Decomposes a DyckWord using its last primitive part
def last_primitive_decomp(dw):
    rev_dw = dw.reverse()
    rev_decomp1, rev_decomp2 = rev_dw.first_return_decomposition()
    if len(rev_decomp1) == 0:
        decomp1 = rev_decomp1
    else:
        decomp1 = rev_decomp1.reverse()
    if len(rev_decomp2) == 0:
        decomp2 = rev_decomp2
    else:
        decomp2 = rev_decomp2.reverse()
    return decomp2, decomp1
```

## 123 Avoiding

```python
def is_123_avoiding_code(code):
    size = len(code)
    for i in range(size):
        for j in range(i+1,size):
            # j is zero indexed so must subtract (j+1)
            if code[i] <= code[j] and code[j] < size - j-1:
                return False

    return True
```

## 213 Avoiding

```python
def is_213_avoiding_code(code):
    size = len(code)
    for i in range(size):
        val = code[i]
        for j in range(i+1,size-val):
            if code[i] > code[j]:
                return False

    return True
```

```python
# counted by large Schroder numbers
def is_nonzero_weak_decr(code):
    size = len(code)
    for i in range(size):
        if code[i] > 0:
            for j in range(i+1,size):
                if code[j] > code[i]:
                    return False

    return True
        
    
```

## 231 Avoiding

```python
def is_231_avoiding_code(code):
    size = len(code)
    for i in range(size):
        val = code[i]
        for j in range(i+1,i+val+1):
            if code[j] > code[i] - (j-i):
                return False

    return True
```

## 321-avoiding alternate (reverse of 123-avoiding)

```python
def is_321_avoiding_code(code):
    size = len(code)
    for i in range(size):
        val = code[i]
        for j in range(i+1,size):
            if i + code[i] >= j + code[j] :
                if code[j] > 0:
                    return False

    return True
```

```python
def is_312_avoiding_code(code):
    size = len(code)
    for i in range(size-1):
        if code[i+1] < code[i] - 1:
            return False

    return True
```

## delta map takes a Dyck path to a 231 avoiding permutation

```python
# this map uses last primitive decomposition
def dw_delta_last(dw):
    #print('>>>processing', dw)
    length = len(dw)
    
    if length == 2:
        return [0]
    else:
        size = length/2
        decomp1, decomp2 = last_primitive_decomp(dw)
        #print('\tdecomp', decomp1, decomp2)

        length2 = len(decomp2)
        # note that size2 is one less than the actual size
        size2 = length2/2

        if length2 == 0:
            code2 = [0]
        else:
            code2 = dw_delta_last(decomp2)
            code2.insert(0, size2)

        if len(decomp1) == 0:
            # primitive
            code = code2
        else:
            code1 = dw_delta_last(decomp1)
            code = code1 + code2
        return code
```

## zeta map takes a Dyck path to a 321-avoiding permutation

```python
# this map uses the last primitive decomposition
def dw_zeta_last(dw):
    #print('>>>processing', dw)
    length = len(dw)
    
    if length == 2:
        return [0]
    else:
        size = length/2
        decomp1, decomp2 = last_primitive_decomp(dw)
        #print('\tdecomp', decomp1, decomp2)

        length2 = len(decomp2)

        if length2 == 0:
            code2 = [0]
        else:
            code2 = dw_zeta_last(decomp2)
            #print('\tcode2', code2)
            code2.insert(0, 1)

        if len(decomp1) == 0:
            # primitive
            code = code2
        else:
            code1 = dw_zeta_last(decomp1)
            for idx in range(len(code1)):
                if code1[idx] > 0:
                    code1[idx] = code1[idx] + 1
            code = code1 + code2
        return code
```

## beta map takes a Dyck path to a 132-avoiding permutation

```python
# this map uses last return decompostion
def dw_beta_last(dw):
    #print('>>>processing', dw)
    length = len(dw)
    
    if length == 2:
        return [0]
    else:
        size = length/2
        decomp1, decomp2 = last_primitive_decomp(dw)
        #print('\tdecomp', decomp1, decomp2)

        length2 = len(decomp2)
        size2 = length2/2

        if length2 == 0:
            code2 = [0]
        else:
            code2 = dw_beta_last(decomp2)
            #print('\tcode1', code1)
            code2.insert(0, size2)

        if len(decomp1) == 0:
            # primitive
            code = code2
        else:
            code1 = dw_beta_last(decomp1)
            #print('\tcode1 before', code1)
            code1 = [x+size2 for x in code1]
            code = code1 + code2
        return code
```

## epsilon map takes a Dyck path to a 312-avoiding permutation
## appears to be the reverse (left to right) of Bandlow-Killpatrick 
## https://www.combinatorics.org/ojs/index.php/eljc/article/view/v8i1r40/pdf
## process their image columnwise to get Lehmer code (instead of rowwise)

```python
# this map uses last primitive decomposition
def dw_epsilon_last(dw):
    #print('>>>processing', dw)
    length = len(dw)
    
    if length == 2:
        return [0]
    else:
        size = length/2
        decomp1, decomp2 = last_primitive_decomp(dw)
        #print('\tdecomp', decomp1, decomp2)

        length2 = len(decomp2)

        if length2 == 0:
            code2 = [0]
        else:
            code2 = dw_epsilon_last(decomp2)
            #print('\tcode2', code2)
            code2 = [x+1 for x in code2]
            code2.append(0)

        if len(decomp1) == 0:
            # primitive
            code = code2
        else:
            code1 = dw_epsilon_last(decomp1)
            code = code1 + code2
        return code
```

## alpha map takes a Dyck path to a 123-avoiding permutation

```python
def dw_alpha_last(dw):
    #print('>>>processing', dw)
    length = len(dw)
    
    if length == 2:
        return [0]
    else:
        size = length/2
        decomp1, decomp2 = last_primitive_decomp(dw)
        #print('\tdecomp', decomp1, decomp2)

        length2 = len(decomp2)
        size2 = length2/2

        if length2 == 0:
            code2 = [0]
        else:
            code2 = dw_alpha_last(decomp2)
            code2 = alpha_one(code2)
            code2 = alpha_two(code2)

        if len(decomp1) == 0:
            # primitive
            code = code2
        else:
            code1= dw_alpha_last(decomp1)
            #print('code1', decomp1, code1, 'code2', decomp2, code2)
            code1 = [x+size2+1 for x in code1]
            code = code1 + code2
        return code

def alpha_one(code):
    return insert_max_nonflush_digit(code)

def alpha_two(code):
    return rearrange_nonflush_descending(code)


#from n-1 lehmer code, I add the "largest not flush" number to the 2nd position, then form a n lehmer code
def is_flush(c):
    n = len(c)
    for i in range(n - 1):
        val = c[i]
        if val == n - i - 1:
            if all(c[j] >= val for j in range(i)):
                return True
    return False

def insert_max_nonflush_digit(p):
    n = len(p) + 1
    for x in reversed(range(n - 1)):
        new_code = [p[0], x] + p[1:]
        if not is_flush(new_code):
            return new_code
    return None

#this is for rearranging. The rule is: I fix all "top" digits, then I rearrange all non-top digits in a decreasing order.
def rearrange_nonflush_descending(p):
    n = len(p)
    fixed_indices = set()
    #fixed_indices.add(0) I found no need, it's always the largest (haven't prove)
    for i in range(n):
        if p[i] == n - i - 1:
            fixed_indices.add(i)
    to_sort = [p[i] for i in range(n) if i not in fixed_indices]
    to_sort_sorted = sorted(to_sort, reverse=True)
    pnew = []
    j = 0
    for i in range(n):
        if i in fixed_indices:
            pnew.append(p[i])
        else:
            pnew.append(to_sort_sorted[j])
            j += 1
    return pnew
```

# Mapping to 132-avoiding permutations using Krattenthaler

## alpha_krat_map is 123-avoiding and 132-avoiding

```python
def alpha_krat_map(size):
    my_map = dict()
    words = DyckWords(size)

    for dw in words:
        alpha = dw_alpha_last(dw)
        perm = permutation.from_lehmer_code(alpha)
        dw_rev = dw.reverse()
        krat = dw_rev.to_132_avoiding_permutation()
        my_map[str(dw)] = [ perm, krat]

    return my_map
```

```python
def beta_krat_map(size):
    my_map = dict()
    words = DyckWords(size)

    for dw in words:
        beta = dw_beta_last(dw)
        perm = permutation.from_lehmer_code(beta)
        dw_rev = dw.reverse()
        krat = dw_rev.to_132_avoiding_permutation()
        my_map[str(dw)] = [ perm, krat]

    return my_map
```
