```python
def get_lehmer(n,pt):
    dw = DyckWords(n)
    lst = []
    for d in dw:
        if pt == a123:
            lst.append(dw_alpha_last(d))
        if pt == a132:
            lst.append(dw_beta_last(d))
        if pt == a213:
            lst.append(dw_gamma_last(d))
        if pt == a231:
            lst.append(dw_delta_last(d))
        if pt == a312:
            lst.append(dw_epsilon_last(d))
        if pt == a321:
            lst.append(dw_zeta_last(d))
    return lst

def get_perm(n,pt):
    dw = DyckWords(n)
    lst = []
    for d in dw:
        if pt == a123:
            lst.append(permutation.from_lehmer_code(dw_alpha_last(d)))
        if pt == a132:
            lst.append(permutation.from_lehmer_code(dw_beta_last(d)))
        if pt == a213:
            lst.append(permutation.from_lehmer_code(dw_gamma_last(d)))
        if pt == a231:
            lst.append(permutation.from_lehmer_code(dw_delta_last(d)))
        if pt == a312:
            lst.append(permutation.from_lehmer_code(dw_epsilon_last(d)))
        if pt == a321:
            lst.append(permutation.from_lehmer_code(dw_zeta_last(d)))
    return lst
        

def first_nonzero_index(lst):
    for i in range(len(lst)):
        if lst[i] != 0:
            return i
    return None

a123 = [1,2,3]
a132 = [1,3,2]
a213 = [2,1,3]
a231 = [2,3,1]
a312 = [3,1,2]
a321 = [3,2,1]
```

```python
def code_inversion_count(code):
    n = len(code)
    count = 0
    for j in range(n):
        for k in range(j + 1, n):
            if code[j] > code[k]:
                count += 1
    return count
```

```python
def count_inv_per(n,pt):
    for i in range(n):
        pa_list = get_perm(i, pt)
        lst = []
        for p in pa_list:
            lst.append(code_inversion_count(p))

        freq = [0] * (max(lst) + 1)
        for x in lst:
            freq[x] += 1

        print("Max_inv:",len(freq)-1, "Min_inv:",first_nonzero_index(freq), "distribution:",freq, sum(freq))

def count_inv_code(n,pt):
    for i in range(n):
        co_list = get_lehmer(i, pt)
        lst2 = []
        for p in co_list:
            lst2.append(code_inversion_count(p))
        freq2 = [0] * (max(lst2) + 1)
        for x in lst2:
            freq2[x] += 1
        print("Max_inv:",len(freq2)-1,"Min_inv:",first_nonzero_index(freq2), "distribution:",freq2,sum(freq2))
        
    return

```

```python
def bw_partition_counts_by_parts(n):
    counts = [0] * (n + 1)
    all_parts = []
    all_list = []

    def backtrack(b_left, w_left, last, cur):
        if b_left == 0 and w_left == 0:
            counts[len(cur)] += 1
            all_parts.append(cur[:])
            return

        for b in range(b_left, 0, -1):
            for w in range(w_left, -1, -1):
                part = (b, w)
                if last is None or part <= last:
                    cur.append(part)
                    backtrack(b_left - b, w_left - w, part, cur)
                    cur.pop()

    for total_b in range(1, n + 1):
        total_w = n - total_b
        backtrack(total_b, total_w, None, [])

    for p in all_parts:
        pbw = ["b" * b + "w" * w for (b, w) in p]
        all_list.append(pbw)

    # print("\ncounts by number of parts:")
    # for k in range(1, n + 1):
    #     print(f"{k} part(s): {counts[k]}")

    return counts, all_list
```

```python
def bw_partition_counts_by_total_b(n):
    counts = [0] * n

    def backtrack(b_left, w_left, last, total_b):
        if b_left == 0 and w_left == 0:
            counts[total_b - 1] += 1
            return
        
        for b in range(b_left, 0, -1):
            for w in range(w_left, -1, -1):
                part = (b, w)
                if last is None or part <= last:
                    backtrack(b_left - b, w_left - w, part, total_b)

    for total_b in range(1, n + 1):
        total_w = n - total_b
        backtrack(total_b, total_w, None, total_b)

    return counts
```

```python
def count_sums_for_inversion_count(n):
    counts = [0] * n

    for p in Permutations(n):
        c = p.to_lehmer_code()
        if code_inversion_count(c) == n-1:
            counts[sum(c)] += 1

    return counts[1:]
```

```python
def grouped_lehmer_codes_sum(n,k):
    groups = {}

    for p in Permutations(n+1):
        c = tuple(p.to_lehmer_code())
        if code_inversion_count(c) == n and sum(c) == k:
            s = sum(c)
            if s not in groups:
                groups[s] = []
            groups[s].append(c)

    #for s in sorted(groups):
        # print("sum =", s)
        # print(groups[s])
        # print()

    return groups
```
